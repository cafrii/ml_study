### 1.10.1. 분류


```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
X = [[0, 0], [1, 1]]
    # X 는 훈련용 샘플.  sparse 또는 dense. shape은 (n_samples, n_features) 형태.
Y = [0, 1]
    # Y 는 정수 값들. shape (n_samples). 훈련 샘플의 클래스 레이블.

clf = DecisionTreeClassifier()
clf = clf.fit(X, Y)
print(type(clf))
```

```python
clf.predict(X=[[2., 2.]], check_input=True)
```

```python
# 특정 클래스를 출력하는 것의 대안으로, 각 클래스의 확률을 예측할 수 있는데, 이는 리프에서 클래스의 훈련 샘플의 일부입니다.
clf.predict_proba([[2., 2.]])
# array([[0., 1.]])  <- 무슨 의미? 잘 이해가 안됨!!
```

```python
iris = sklearn.datasets.load_iris()
X, y = iris.data, iris.target

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.plot_tree(clf)
```

```python
import graphviz

dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
# graph.render("iris")  # 현재 폴더에 iris.pdf 를 생성하므로, 실행하진 말자.
```

```python
# Export a decision tree in DOT format.
dot_data = tree.export_graphviz(clf, out_file=None,
                     feature_names=iris.feature_names,
                     class_names=iris.target_names,
                     filled=True, rounded=True,
                     special_characters=True)
graph = graphviz.Source(dot_data)
print(type(graph))
graph
```

```python
# 또는 export_text 함수를 사용하여 트리를 텍스트 형식으로 내보낼 수도 있습니다.
# 이 방법은 외부 라이브러리를 설치할 필요가 없고 더 간결합니다.

from sklearn.tree import export_text

iris = sklearn.datasets.load_iris()
decision_tree = DecisionTreeClassifier(random_state=0, max_depth=2)
decision_tree = decision_tree.fit(iris.data, iris.target)
r = export_text(decision_tree, feature_names=iris['feature_names'])
print(r)
```
